```python
# alt.Chart.encode?
# alt.condition
# alt.Chart.configure_title?
# alt.Chart.configure_text?
# alt.Axis?
```

```python
NASA_plot = alt.Chart(
    prep_df,
    title={
        "text": ['Annual Global Surface Temperature', 'Relative to Late 19th Century Average'],
    }
).mark_point().encode(
    x = alt.X(
        field='year_datetime',
        type='temporal',
        title = None,
        axis=alt.Axis(
            values=list(set(prep_df.decade))
        ),
    ),
    y = alt.Y(
        field ='value',
        title = None,
        type='quantitative',
        axis=alt.Axis(
            gridWidth=alt.condition('datum.value == 0', alt.value(1.5), alt.value(.5)),
            gridColor=alt.condition('datum.value == 0', alt.value('#696969'),  alt.value('#B8B8B8')),
            values=y_axis,
        ),
    ),
    tooltip = [
        alt.Tooltip(title='Year', field='year_datetime', type='temporal', format='%Y'),
    ]
)

text_OX = alt.Chart({'values':[{'x': 2010, 'y': 0}]}).mark_text(
    text='HOTTER THAN THE\n1880-1899 AVERAGE \n\nCOLDER',
    fontSize=12,
    fontWeight=900,
    color='#696969',
    dx=300,
    dy=-27,
    lineBreak='\n'
).encode(
    x='x:T',
    y='y:Q'
)



alt.layer(NASA_plot, text_OX).configure_point(
    color='black',
    fill='darkorange',
    strokeWidth=1,
).configure_view(
    strokeWidth = 1,
    strokeOpacity=0,
).configure_axis(
    ticks = False,
    domain = False,
    labelFontSize=12,
    labelColor='#B8B8B8'
).configure_title(
    anchor='start',
    dx = 50,
    dy = 80,
    fontSize=16,
).configure_text(
    align='right',
    fontSize=12,
    fontWeight=900
).properties(
    width = 900,
    height = 500,
    background = '#FFFFFF',
    padding = 25
)
```
